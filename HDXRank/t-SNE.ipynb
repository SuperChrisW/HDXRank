```python
# t-SNE plot of pep embeddings
import numpy as np
from torch.nn.functional import pad
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster

# numpy, torch, sklearn, bioinfokit required
```

```python
def load_data():
    # load your data here
    #e.g a random generated data

    X_data = np.random.rand(100, 10)
    Y_data = np.random.rand(100)
    return X_data, Y_data

    # if sequence is in different length pad them to the same length
    # max_len = max([len(seq) for seq in X_data])
    # X_data = pad(X_data, (0, max_len - len(seq)), 'constant', 0)
    pass

X_data, Y_data = load_data() # X_data is a numpy array of shape (n_samples, n_features), Y_data is a list of labels
# classify Y labels into 5 classes
Y_bin = np.digitize(Y_data, [0.2, 0.4, 0.6, 0.8])
print(X_data.shape)
print(Y_bin.shape)
```

```python
tsne_em = TSNE(n_components=3, perplexity=30.0, n_iter=1000, early_exaggeration=12, learning_rate=368, verbose=1).fit_transform(X_data)

# perplexity usually range between 10-100
# n_iter usually between 1000-10000
# tutorial: https://www.datacamp.com/tutorial/introduction-t-sne
```

```python
cluster.tsneplot(score=tsne_em, colorlist=Y_bin, legendpos='upper right', 
colordot=('#407DD7', '#40B6BD', '#A982C0', '#EEB2E2', '#D46853', '#1A2B47'), valphadot=0.8, show=True,
legendanchor=(1.35, 1), figname = 'someplace_to_save') # to save the plot, let show = False
```
